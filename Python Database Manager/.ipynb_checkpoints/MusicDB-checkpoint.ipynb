```python
import sqlite3
def connect():
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("CREATE TABLE IF NOT EXISTS artist (id INTEGER PRIMARY KEY, name text, gender INTEGER)")
    cur.execute("CREATE TABLE IF NOT EXISTS gender (id INTEGER PRIMARY KEY, name text)")
    cur.execute("CREATE TABLE IF NOT EXISTS type (id INTEGER PRIMARY KEY, name text)")
    cur.execute("CREATE TABLE IF NOT EXISTS language (id INTEGER PRIMARY KEY, name text)")
    cur.execute("CREATE TABLE IF NOT EXISTS genre (id INTEGER PRIMARY KEY, name text)")
    cur.execute("CREATE TABLE IF NOT EXISTS album (id INTEGER PRIMARY KEY, name text, type INTEGER, artist INTEGER)")
    cur.execute("CREATE TABLE IF NOT EXISTS song (id INTEGER PRIMARY KEY, title text, language INTEGER, album INTEGER)")
    cur.execute("CREATE TABLE IF NOT EXISTS video (id INTEGER PRIMARY KEY, name text, genre INTEGER, artist INTEGER)")
    conn.commit()
    conn.close()
    
def insert_artist(name, gender):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO artist VALUES(NULL, ?, ?)", (name, gender))
    conn.commit()
    conn.close()
    
def view_artist():
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM artist")
    rows=cur.fetchall()
    conn.close()
    return rows

def search_artist(name="", gender=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM artist WHERE name=? OR gender=?", (name, gender))
    rows=cur.fetchall()
    conn.close()
    return rows

def delete_artist(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    if len(view_album(id)) == 0 and len(view_video(id)) == 0:
        cur.execute("DELETE FROM artist WHERE id=?", (id,))
    conn.commit()
    conn.close()

def update_artist(id, name="", gender=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("UPDATE artist SET name=?, gender=? WHERE id=?", (name, gender, id))
    conn.commit()
    conn.close()


def insert_album(artist, name, typ):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO album VALUES(NULL, ?, ?, ?)", (name, typ, artist))
    conn.commit()
    conn.close()
    
def view_album(artist):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM album WHERE artist=?", (artist,))
    rows=cur.fetchall()
    conn.close()
    return rows

def search_album(artist, name="", typ=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM album WHERE (name=? OR type=?) AND artist=?", (name, typ, artist))
    rows=cur.fetchall()
    conn.close()
    return rows

def delete_album(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM album WHERE id=?", (id,))
    conn.commit()
    conn.close()

def update_album(id, artist, name="", typ=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("UPDATE album SET name=?, type=?, artist=? WHERE id=?", (name, typ, artist, id))
    conn.commit()
    conn.close()
  



def insert_song(album, title, language):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO song VALUES(NULL, ?, ?, ?)", (title, language, album))
    conn.commit()
    conn.close()
    
def view_song(album):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM song WHERE album=?", (album,))
    rows=cur.fetchall()
    conn.close()
    return rows

def search_song(album, title="", language=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM song WHERE (title=? OR language=?) AND album=?", (title, language, album))
    rows=cur.fetchall()
    conn.close()
    return rows

def delete_song(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM song WHERE id=?", (id,))
    conn.commit()
    conn.close()

def update_song(id, album, title="", language=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("UPDATE song SET title=?, language=?, album=? WHERE id=?", (title, language, album, id))
    conn.commit()
    conn.close()


def insert_video(artist, name, genre):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO video VALUES(NULL, ?, ?, ?)", (name, genre, artist))
    conn.commit()
    conn.close()

def view_video(artist):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM video WHERE artist=?", (artist,))
    rows=cur.fetchall()
    conn.close()
    return rows

def search_video(artist, name="", genre=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM video WHERE (name=? OR genre=?) AND artist=?", (name, genre, artist))
    rows=cur.fetchall()
    conn.close()
    return rows

def delete_video(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM video WHERE id=?", (id,))
    conn.commit()
    conn.close()

def update_video(id, artist, name="", genre=""):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("UPDATE video SET name=?, genre=?, artist=? WHERE id=?", (name, genre, artist, id))
    conn.commit()
    conn.close()





    
    
def get_gender(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM gender WHERE id=?", (id,))
    rows=cur.fetchall()
    conn.close()
    return rows

def view_gender():
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM gender")
    rows=cur.fetchall()
    conn.close()
    return rows
    
def insert_gender(name):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO gender VALUES(NULL, ?)", (name,))
    conn.commit()
    conn.close()
    
def delete_gender(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM gender WHERE id=?", (id,))
    conn.commit()
    conn.close()


def get_type(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM type WHERE id=?", (id,))
    rows=cur.fetchall()
    conn.close()
    return rows

def view_type():
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM type")
    rows=cur.fetchall()
    conn.close()
    return rows
    
def insert_type(name):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO type VALUES(NULL, ?)", (name,))
    conn.commit()
    conn.close()
    
def delete_type(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM type WHERE id=?", (id,))
    conn.commit()
    conn.close()

    
    

    
def get_language(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM language WHERE id=?", (id,))
    rows=cur.fetchall()
    conn.close()
    return rows

def view_language():
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM language")
    rows=cur.fetchall()
    conn.close()
    return rows
    
def insert_language(name):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO language VALUES(NULL, ?)", (name,))
    conn.commit()
    conn.close()
    
def delete_language(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM language WHERE id=?", (id,))
    conn.commit()
    conn.close()


def get_genre(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM genre WHERE id=?", (id,))
    rows=cur.fetchall()
    conn.close()
    return rows

def view_genre():
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("SELECT * FROM genre")
    rows=cur.fetchall()
    conn.close()
    return rows
    
def insert_genre(name):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("INSERT INTO genre VALUES(NULL, ?)", (name,))
    conn.commit()
    conn.close()
    
def delete_genre(id):
    conn=sqlite3.connect("music.db")
    cur=conn.cursor()
    cur.execute("DELETE FROM genre WHERE id=?", (id,))
    conn.commit()
    conn.close()



    
    
connect()
# insert("The Sun", 2)
# delete(2)
delete_video(1)
# print(get_type(1))
print(view_video(1))
# print(search(name='The Moon'))
```

```python
from tkinter import *


# Here, we are creating our class, Window, and inheriting from the Frame
# class. Frame is a class from the tkinter module. (see Lib/tkinter/__init__)
class Window(Frame):

    # Define settings upon initialization. Here you can specify
    def __init__(self, master=None):
        
        # parameters that you want to send through the Frame class. 
        Frame.__init__(self, master)   

        #reference to the master widget, which is the tk window                 
        self.master = master

        #with that, we want to then run init_window, which doesn't yet exist
        self.init_window()

    #Creation of init_window
    def init_window(self):

        # changing the title of our master widget      
        self.master.title("GUI")

        # allowing the widget to take the full space of the root window
        self.pack(fill=BOTH, expand=1)

        # creating a button instance
        quitButton = Button(self, text="Exit",command=self.client_exit)

        # placing the button on my window
        quitButton.place(x=0, y=0)

       

    def client_exit(self):
        exit()

# root window created. Here, that would be the only window, but
# you can later have windows within windows.
root = Tk()

root.geometry("400x300")

#creation of an instance
app = Window(root)

#mainloop 
root.mainloop() 
```
